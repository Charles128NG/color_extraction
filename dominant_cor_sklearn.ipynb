# dominant color extraction using k means clustering

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/home/charles/Pictures/b1.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
print(img.shape)
orig = img
```

```python
all_pixels = img.reshape((265*300 ,3))
all_pixels.shape
```

```python
from sklearn.cluster import KMeans
```

# training model to extract 5 dominant colors from our picture

```python
dom_cols = 3
km = KMeans(n_clusters = dom_cols)
km.fit(all_pixels)
```

```python
centers = km.cluster_centers_
centers = np.array(centers,dtype = 'uint8')
centers
```

# plot all colors

```python
i = 1
colors = [];
for each_col in centers:
    plt.subplot(1,3,i)
 
    i+=1
    
    colors.append(each_col)
    
    a =np.zeros((100,100,3),dtype = 'uint8')
    a[:,:,:] = each_col
    plt.imshow(a)
    plt.axis("off")
plt.show()
```

# segmenting the image

```python
new_img = np.zeros((79500, 3),dtype = 'uint8')
a = km.labels_
for ix in range(new_img.shape[0]):
    new_img[ix] = colors[a[ix]]
    
new_img = new_img.reshape(orig.shape)
plt.imshow(new_img)
plt.show()
```
